```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('B6_bh.csv', header=None)
```

```python
def cost(CepRs, REs, Co=0.0, TM=0.0, B=6.0):
    res = 140 * REs / (140 * REs + 10.8 * B + 56*(100-REs-B))
    lace = CepRs * res
    nd = res - lace
    cost = lace*37 + nd*480
    return [cost]
```

```python
list_1 = []
ceprs = np.linspace(0, 1, 21)
res = np.linspace(6, 18, 21)

for i in ceprs:
    for j in res:
        list_1.append(cost(i, j))

a = np.array(list_1).reshape(len(ceprs), len(res))
cost_1 = pd.DataFrame(a)
```

```python
perfor_cost = df / cost_1 * 6.6
p_c = pd.DataFrame(perfor_cost.values, index=ceprs, columns=res)
p_c.to_csv('pc_bh.csv')
```
